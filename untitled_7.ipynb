```python
from sklearn.datasets import load_digits
import numpy as np
```

```python
digits = load_digits()
X= digits.data
y= digits.target
```

```python
import matplotlib.pyplot as plt

plt.imshow(X[33].reshape(8, 8), cmap='gray')
plt.title(f"Label: {y[33]}")
plt.show()
```

```python
X/=16.0
```

```python
def one_hot(y, num_classes=10):
  return np.eye(num_classes)[y]

y_onehot = one_hot(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_onehot, test_size=0.2, random_state=42
)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def softmax(z):
    # 1. Subtract max for stability: e^(z - max)
    # axis=1 handles batches; keepdims=True ensures shapes align for subtraction
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))

    # 2. Divide by the sum of exponents
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)
```

```python
def he_normal(n_in, n_out):
    std = np.sqrt(2.0 / n_in)
    return np.random.randn(n_in, n_out) * std
```

```python
def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15  # To avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred)) / y_true.shape[0]
```

```python
W1 = he_normal(64, 128)
b1 = np.zeros((1,128))
z1 = X_train@W1+b1
l1 = relu(z1)

W2 = he_normal(128,64)
b2 = np.zeros((1,64))
z2 = l1@W2+b2
l2 = relu(z2)

W3 = he_normal(64,32)
b3 = np.zeros((1,32))
z3 = l2@W3+b3
l3 = relu(z3)

W4 = he_normal(32,10)
b4 = np.zeros((1,10))
z4 = l3@W4+b4
l4 = softmax(z4)
```

```python
# Hyperparameters
epochs = 1000
learning_rate = 0.01
m = X_train.shape[0]  # Number of training samples (used for averaging gradients)

# Track loss history for debugging
loss_history = []

for epoch in range(epochs):
    # --- 1. FORWARD PASS ---
    z1 = X_train @ W1 + b1
    l1 = relu(z1)

    z2 = l1 @ W2 + b2
    l2 = relu(z2)

    z3 = l2 @ W3 + b3
    l3 = relu(z3)

    z4 = l3 @ W4 + b4
    l4 = softmax(z4)  # This is our y_pred

    # Compute and record the loss
    loss = cross_entropy_loss(y_train, l4)
    loss_history.append(loss)

    # --- 2. BACKWARD PASS (Backpropagation) ---
    # Output Layer (Layer 4)
    # Derivative of Cross-Entropy + Softmax simplifies beautifully to: (y_pred - y_true)
    dz4 = l4 - y_train
    dW4 = (l3.T @ dz4) / m
    db4 = np.sum(dz4, axis=0, keepdims=True) / m

    # Hidden Layer 3
    # Derivative of ReLU is 1 if z > 0, else 0
    dl3 = dz4 @ W4.T
    dz3 = dl3 * (z3 > 0)
    dW3 = (l2.T @ dz3) / m
    db3 = np.sum(dz3, axis=0, keepdims=True) / m

    # Hidden Layer 2
    dl2 = dz3 @ W3.T
    dz2 = dl2 * (z2 > 0)
    dW2 = (l1.T @ dz2) / m
    db2 = np.sum(dz2, axis=0, keepdims=True) / m

    # Hidden Layer 1
    dl1 = dz2 @ W2.T
    dz1 = dl1 * (z1 > 0)
    dW1 = (X_train.T @ dz1) / m
    db1 = np.sum(dz1, axis=0, keepdims=True) / m

    # --- 3. PARAMETER UPDATES (Gradient Descent) ---
    W4 -= learning_rate * dW4
    b4 -= learning_rate * db4

    W3 -= learning_rate * dW3
    b3 -= learning_rate * db3

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    # Print progress every 100 epochs
    if epoch % 100 == 0 or epoch == epochs - 1:
        print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.imshow(X_test[2].reshape(8,8), cmap='gray')
plt.show()

z1 = X_test[2] @ W1 + b1
l1 = relu(z1)

z2 = l1 @ W2 + b2
l2 = relu(z2)

z3 = l2 @ W3 + b3
l3 = relu(z3)

z4 = l3 @ W4 + b4
l4 = softmax(z4)

np.argmax(l4) == np.argmax(y_test[2])
```

```python
corrects = 0
total = len(y_test)
for i in range(total):
  z1 = X_test[2] @ W1 + b1
  l1 = relu(z1)

  z2 = l1 @ W2 + b2
  l2 = relu(z2)

  z3 = l2 @ W3 + b3
  l3 = relu(z3)

  z4 = l3 @ W4 + b4
  l4 = softmax(z4)

  if np.argmax(l4) == np.argmax(y_test[2]):
    corrects+=1

corrects/total * 100
```
