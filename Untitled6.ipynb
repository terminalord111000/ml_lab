```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader

```

```python
train_data = datasets.MNIST(
    root = 'data',
    train = True,
    transform = ToTensor(),
    download = True
)
```

```python
test_data = datasets.MNIST(
    root = 'data',
    train = False,
    transform = ToTensor(),
    download = True
)
```

```python
loaders = {
    'train': DataLoader(train_data,
                        batch_size=100,
                        shuffle=True,
                        num_workers=0),

    'test': DataLoader(test_data,
                       batch_size=100,
                       shuffle=False,
                       num_workers=0)
}
```

```python
'''
so what I have learnt
the first class is basically a blueprint of the input for
each hidden layer and the what is outputs
eg. fc3(input_nuerons from pervious layer, num of nuerons in
the layer that is input for another layer so effectively an output)

then the forward method
first now flattens the matrix to match the design specifications
in the constructor
then you take each hidden layer, pass the actual data into that layer
then you apply activation to that layer. till you reach the output
'''

class MLP(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(28*28,128)
    self.fc2 = nn.Linear(128,64)
    self.fc3 = nn.Linear(64,10)

  def forward(self, x):
    x = x.view(x.size(0), -1)

    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.fc3(x)

    return x
```

```python
'''
- you define the device
- you instatiate the model to run on that device
- setup your loss
- setup your optimizer
'''

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MLP().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 5
train_loader = loaders['train']

for epoch in range(num_epochs):
  model.train()
  running_loss = 0.0

  for images, labels in train_loader:
    images, labels = images.to(device), labels.to(device)
    outputs = model(images)
    loss = criterion(outputs, labels)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    running_loss += loss.item()

  print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}")

```

```python
'''
so the loop. goes through the image batch by batch. then for each batch.
we check the output of the model and we pick the output with the higherst
number add the batch number basically to a total, then we count the number
of correct ones by comparing predicted to the labels. then we sum them up.
then we calculate percentage. then we move to the next batch and do thesame thing.
'''

test_loader = loaders['test']
model.eval()
total = 0
correct = 0

with torch.no_grad():
  for images,labels in test_loader:
    images, labels = images.to(device), labels.to(device)

    outputs = model(images)
    _, predicted = torch.max(outputs,1)
    total+=labels.size(0)
    correct+=(predicted==labels).sum().item()

  print(f"Test Accuracy: {100 * correct / total:.2f}%")

```

```python
'''
TEST YOUR MODEL ON A CUSTOM IMAGE
Just paste this code block into a new cell in your notebook
'''

from PIL import Image
import matplotlib.pyplot as plt

# STEP 1: Load your image
image_path = '/content/10.png'  # Change this to your image path
img = Image.open(image_path)

# Convert to grayscale if it's color
if img.mode != 'L':
    img = img.convert('L')

# Resize to 28x28 (MNIST size)
img = img.resize((28, 28))

# Display the image to see what you're testing
plt.imshow(img, cmap='gray')
plt.title("Your Test Image")
plt.show()

# STEP 2: Convert image to tensor
img_tensor = ToTensor()(img)
img_tensor = img_tensor.unsqueeze(0)  # Add batch dimension
img_tensor = img_tensor.to(device)

# STEP 3: Use the model to predict
model.eval()
with torch.no_grad():
    output = model(img_tensor)
    probabilities = F.softmax(output, dim=1)
    predicted_digit = torch.argmax(output, dim=1).item()
    confidence = probabilities[0][predicted_digit].item()

# STEP 4: Show the results
print(f"\n🎯 Predicted Digit: {predicted_digit}")
print(f"📊 Confidence: {confidence * 100:.2f}%")
print("\nAll probabilities:")
for digit in range(10):
    prob = probabilities[0][digit].item()
    print(f"  Digit {digit}: {prob * 100:.2f}%")
```
